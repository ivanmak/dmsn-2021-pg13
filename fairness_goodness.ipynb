````python
from temporal_networkx import TemporalDiGraph
import pandas as pd
import numpy as np
import networkx as nx
import seaborn as sb
import matplotlib.pyplot as plt
import math
````

````python
temporal_graph = TemporalDiGraph(pd.read_csv('./soc-sign-bitcoinotc.csv', header=None))
````

If the method get_DiGraph is run with run_REV2 set to True, it will return a directed graph
with the additional attributes at the nodes - fairness and goodness, which are generated
using REV2 algorithm

````python
G = temporal_graph.get_DiGraph('2012-01', run_REV2=True)
````

````python
G.nodes[1576]
````

The original draw_network() method still works as usual

````python
temporal_graph.draw_network('2012-01')
````

But if draw_network() is called with ```graph_type``` set to 'rev2',
it will draw the network by having node colour being relative to the 
goodness score of the nodes, and the node size being relative to the
fairness score of the nodes.

````python
temporal_graph.draw_network('2012-01', graph_type='rev2')
````
